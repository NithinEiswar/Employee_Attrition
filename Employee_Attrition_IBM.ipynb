```python
#Remove unneeded columns

#Remove the column EmployeeNumber
df = df.drop('EmployeeNumber', axis = 1) # A number assignment 
#Remove the column Behaviour
df = df.drop('Behaviour', axis = 1) #Contains only value 1
#Remove the column EmployeeCount
df = df.drop('Id', axis = 1) #A number assignment
```

```python
df.shape
```

```python
y=df['Attrition']
```

```python
#Visualize the correlation
plt.figure(figsize=(14,14))  #14in by 14in
sns.heatmap(df.corr(), annot=True, fmt='.0%')
```

```python
#Remove YearsWithCurrManager as it is highly correlated
df=df.drop(['YearsWithCurrManager'],axis=1)
```

```python
#Transform non-numeric columns into numerical columns
from sklearn.preprocessing import LabelEncoder

for column in df.columns:
        if df[column].dtype == np.number:
            continue
        df[column] = LabelEncoder().fit_transform(df[column])
```

```python
x=df.drop(['Attrition'],axis=1)
```

```python
# Split the dataset into 75% Training set and 25% Testing set
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(x, y, test_size = 0.25, random_state = 0)
```

```python
#Grid Search
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV
clf = LogisticRegression(max_iter=10000,dual=False)

grid={"C":np.logspace(-4,4,20), "penalty":["l2"]}# l1 lasso l2 ridge
grid_clf_acc = GridSearchCV(clf, param_grid = grid,scoring = 'roc_auc')
grid_clf_acc.fit(X_train, Y_train)

y_pred=grid_clf_acc.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score,roc_auc_score

print(confusion_matrix(Y_test,y_pred))
print(classification_report(Y_test,y_pred))
print(accuracy_score(Y_test, y_pred))
print(roc_auc_score(Y_test, y_pred))
```
